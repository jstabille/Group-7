```python
sql = """
SELECT "table_name","column_name", "data_type", "table_schema"
FROM INFORMATION_SCHEMA.COLUMNS
WHERE "table_schema" = 'public'
ORDER BY table_name  
"""
pd.read_sql(sql, con=connection)
```

```python
sql_USDisaster = """ SELECT * FROM public."USDisaster" ORDER BY declaration_date DESC LIMIT 10 """
pd.read_sql(sql_USDisaster, con=connection)
```
